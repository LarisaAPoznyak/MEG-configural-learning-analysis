```python
#day 2
for subj in subjs:

    df = pd.DataFrame(columns=['subject', 'run',  'MISC001'])
    
    for i in range(1, 4):
        prot = subj[2]###
        subject = subj[0]
        path = f"E:/MSUPE/piansrann/meg/{subj[0]}/{subj[1]}/" ###
        run = i
        day = '2'
        file = f'{subject}_{prot[0]}_{day}_run{run}_raw_tsss.fif'  #####
        filename = path+file
        raw = mne.io.read_raw_fif(filename, preload=True)
        pivot_channels = mne.pick_types(raw.info, meg=False, eeg=False, stim=True, eog=False, misc=True)
        channels_of_interest = ['MISC001']
        channels_to_keep = [ch for ch in pivot_channels if raw.info['ch_names'][ch] in channels_of_interest]

        # Extract the data for the selected channels
        for_scr = raw.copy().pick_channels([raw.info['ch_names'][ch] for ch in channels_to_keep])
        #events
        events = mne.find_events(raw, min_duration=0, shortest_event = 0 )
        exclude=[1, 16, 32, 64, 128] #exclusion of event we do not need for further analysis ###############
        events = mne.pick_events(events, include=None, exclude=exclude, step=False)
        events_clean =[]
        tok = [130, 140, 132, 134, 136, 138]
        for i in range(0, len(events[:, 2])-3):
            if events[:, 2][i] == events[:, 2][i+2] or events[:, 2][i+2] in tok:
                events_clean.append(events[i])
        events_clean.append(events[len(events[:, 2])-3])
        events_clean = np.array(events_clean)
        if prot == b:
            for i in range(len(events_clean)):
                if events_clean[:, 2][i] == 12:
                    events_clean[:, 2][i] = 13
                if events_clean[:, 2][i] == 4:
                    events_clean[:, 2][i] = 5

        #epochs 
        epochs1 = mne.Epochs(for_scr, events_clean, picks=['MISC001'], event_id = prot[1], tmin=-1, tmax=0, baseline=None, detrend = None, preload = True, reject_by_annotation = True,  on_missing='ignore')
        df1 = epochs1.to_data_frame()
        result1 = df1.groupby(["epoch", "condition"])['MISC001'].mean()
        result1 = result1.droplevel('epoch')
        result1 = result1.to_frame()
        result1.insert (0, 'run', run)
        result1.insert (0, 'subject', subject)
        result1.insert (0, 'time', '-1 - 0')
        result1.insert (0, 'prot', prot[0])
        
        epochs2 = mne.Epochs(for_scr, events_clean, picks=['MISC001'], event_id = prot[1], tmin=0.970, tmax=1.97, baseline=None, detrend = None, preload = True, reject_by_annotation = True,  on_missing='ignore')
        df2 = epochs2.to_data_frame()
        result2 = df2.groupby(["epoch", "condition"])['MISC001'].mean()
        result2 = result2.to_frame()
        result2 = result2.droplevel('epoch')
        result2.insert (0, 'run', run)
        result2.insert (0, 'subject', subject)
        result2.insert (0, 'time', '0.97 - 1.97')
        result2.insert (0, 'prot', prot[0])

        epochs3 = mne.Epochs(for_scr, events_clean, picks=['MISC001'], event_id = prot[1], tmin=2.94, tmax=3.94, baseline=None, detrend = None, preload = True, reject_by_annotation = True,  on_missing='ignore')
        df3 = epochs3.to_data_frame()
        result3 = df3.groupby(["epoch", "condition"])['MISC001'].mean()
        result3 = result3.to_frame()
        result3 = result3.droplevel('epoch')
        result3.insert (0, 'run', run)
        result3.insert (0, 'subject', subject)
        result3.insert (0, 'time', '2.94 - 3.94')
        result3.insert (0, 'prot', prot[0])

        test = pd.concat([result1, result2, result3], axis=0)
        df = pd.concat([df, test], axis=0)
        df.to_csv(f'E:/MSUPE/piansrann/tables_day2/{subject}.csv')

```
